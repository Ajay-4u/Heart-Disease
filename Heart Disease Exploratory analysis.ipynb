# Profile of people likely to develop a heart disease from my analysis

From my analysis Age, cp (chest pain experinced), Cholestorol, Thalach (maximum heart rate achieved), Oldpeak (ST depression induced by exercise relative to rest), ca (number of major vessels), and thal (a blood disorder called thalassemia) have proven to be contributing factors to heart failure.

People with heart failures tend to;
- be of ages 29-54 
- experience the atypical angina, non-anginal pain, or asymptomatic chest pain 
- have cholestorol levels of 126-254
- have a thalach level from 148-202 
- have an oldpeak level ranging from 0-1.6 
- have zero(0) major vessels (ca)
- and have the blood disorder called thalassemia (thal)
